* B√∫squeda por palabras clave: Extrae palabras clave de la pregunta del usuario y busca coincidencias en las preguntas almacenadas.

* Similitud del coseno: Si has representado las preguntas como vectores (por ejemplo, usando TF-IDF o word embeddings), puedes usar la similitud del coseno para medir la distancia entre las preguntas.

* Word embeddings: Utiliza modelos de word embeddings como Word2Vec o BERT para obtener representaciones sem√°nticas de las preguntas y las consultas del usuario.

```python
tu_diccionario = {
   "hola": "¬°Hola! ¬øEn qu√© puedo ayudarte?",
   "adi√≥s": "Hasta luego. ¬°Que tengas un buen d√≠a!",
   "informaci√≥n": "¬øQu√© tipo de informaci√≥n est√°s buscando?",
   # Agrega m√°s entradas de diccionario seg√∫n tus necesidades
}
```

```python
def responder_pregunta(pregunta):
    # Usar split para un procesamiento simple sin nlp, ya que el objeto nlp se define m√°s tarde
    pregunta_procesada = pregunta.lower().split()
    respuesta = "Lo siento, no entiendo tu pregunta."

    # Busca una coincidencia en el diccionario
    for palabra in pregunta_procesada:
        # regresa la primer coincidencia que encuentra
        if palabra in tu_diccionario:
            respuesta = tu_diccionario[palabra]
            break

    return respuesta
```

## B√∫squeda por similitud

Para los chatbots basados ‚Äã‚Äãen recuperaci√≥n, es com√∫n utilizar bolsas de palabras (bag of words) o tf-idf para calcular la similitud de intenciones.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Datos de ejemplo
preguntas = ["¬øQu√© es el aprendizaje autom√°tico?",
             "¬øC√≥mo funciona la regresi√≥n lineal?"]
respuestas = ["El aprendizaje autom√°tico es una rama de la inteligencia artificial...",
              "La regresi√≥n lineal es un m√©todo de modelado..."]

# Vectorizaci√≥n con TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(preguntas)

# Funci√≥n para encontrar la mejor coincidencia
def responder_pregunta(consulta_usuario):
    consulta_vec = vectorizer.transform([consulta_usuario])
    similitudes = cosine_similarity(consulta_vec, tfidf_matrix).flatten()
    print(similitudes)
    indice_mejor_coincidencia = similitudes.argmax()
    print(indice_mejor_coincidencia)
    return respuestas[indice_mejor_coincidencia]
```

```python

# Ejemplo de consulta
consulta = "¬øQu√© es la regresi√≥n lineal?"
print(responder_pregunta(consulta))
```

###  2) Crear el chatbot utilizando TFIDF y similitud del coseno. (1 punto)



##### Introducci√≥n al Script

Para este punto, implementar√© el **primer chatbot**.  
Utilizar√© la biblioteca **scikit-learn** para dos tareas clave:

- `TfidfVectorizer`: Convertir√° todas las preguntas de mi dataset en una **matriz de vectores num√©ricos**.  
  TF-IDF mide la **importancia de una palabra** en un documento (pregunta) en relaci√≥n con el corpus completo (todas las preguntas).

- `cosine_similarity`: Calcular√° la **similitud del coseno** entre el vector de la consulta del usuario y los vectores de todas las preguntas del dataset.

La funci√≥n `responder_tfidf` encapsular√° este proceso:  
recibir√° una **consulta**, la **vectorizar√°**, encontrar√° el **√≠ndice de la pregunta m√°s similar** (`argmax`) y devolver√° la **respuesta** y la **pregunta original** correspondiente a ese √≠ndice (para poder verificarla en el punto 4).


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 1. Inicializar el Vectorizador TF-IDF
# Usar√© 'spanish' para las stopwords (palabras comunes como 'el', 'la', 'de')
# FIX: Removed stop_words='spanish' as it's not directly supported by TfidfVectorizer.
vectorizer_tfidf = TfidfVectorizer()

# 2. Ajustar y transformar el dataset de preguntas
# Esto crea la matriz TF-IDF (nuestra base de conocimiento)
tfidf_matrix = vectorizer_tfidf.fit_transform(preguntas)

print("Forma de la matriz TF-IDF:", tfidf_matrix.shape)

# 3. Funci√≥n para responder
def responder_tfidf(consulta_usuario):
    """
    Encuentra la mejor respuesta usando TF-IDF y similitud del coseno.
    """
    # 1. Vectorizar la consulta del usuario usando el MISMO vectorizador
    # (Usamos .transform() no .fit_transform())
    consulta_vec = vectorizer_tfidf.transform([consulta_usuario])

    # 2. Calcular la similitud del coseno
    # Compara el vector de la consulta (1, N) con la matriz de preguntas (50, N)
    similitudes = cosine_similarity(consulta_vec, tfidf_matrix).flatten()

    # 3. Encontrar el √≠ndice de la mayor similitud
    indice_mejor_coincidencia = np.argmax(similitudes)

    # 4. Devolver la respuesta y la pregunta original (para verificaci√≥n)
    if similitudes[indice_mejor_coincidencia] > 0: # Umbral simple para evitar coincidencias nulas
        return respuestas[indice_mejor_coincidencia], preguntas[indice_mejor_coincidencia]
    else:
        return "Lo siento, no entiendo tu pregunta.", None

# Prueba r√°pida (opcional)
# consulta_prueba = "¬øcu√°nto sale el env√≠o?"
# respuesta, pregunta_match = responder_tfidf(consulta_prueba)
# print(f"Consulta: {consulta_prueba}")
# print(f"Respuesta: {respuesta}")
# print(f"Pregunta Original: {pregunta_match}")
```

##### Observaci√≥n del Script

El `vectorizer_tfidf` ha sido **entrenado (ajustado)** con las **50 preguntas** de mi dataset.  
La salida `(50, N)` (donde `N` es el tama√±o del vocabulario) confirma que tenemos **50 vectores de preguntas**.  
La funci√≥n `responder_tfidf` est√° lista.  

Es importante destacar que este modelo es **superficial**: solo compara **palabras clave**.  
Si pregunto *"¬øcu√°ndo recibo mi compra?"* y en el dataset tengo *"¬øcu√°nto tarda en llegar mi pedido?"*,  
el **TF-IDF** podr√≠a tener problemas si las palabras clave **no coinciden lo suficiente**.

### 3) Crear otro chatbot utilizando embeddings. Indique cu√°l embedding (1 punto) pre-entrenado eligi√≥.



##### Introducci√≥n al Script

Para el **segundo chatbot**, usar√© **Embeddings pre-entrenados**, que capturan el **significado sem√°ntico** de las palabras.  
He elegido usar **spaCy** con el modelo `es_core_news_sm`.

**Elecci√≥n del Embedding:** `es_core_news_sm`  
**Motivo:** Es un modelo **peque√±o y eficiente** para espa√±ol, ideal para este ejercicio.  
Ya incluye **vectores de palabras (embeddings)** pre-entrenados.

**El proceso ser√°:**

1. **Cargar** el modelo `spaCy`.  
2. **Obtener** el **vector de documento (embedding)** para cada una de mis 50 preguntas.  
   Un vector de documento en spaCy (para modelos *sm*) es el **promedio de los vectores de las palabras** que lo componen.  
3. **Almacenar√©** estos 50 vectores en una **matriz NumPy**.  
4. **Crear√©** una funci√≥n `responder_spacy` que:  
   a. Obtenga el **vector de la consulta del usuario**.  
   b. Calcule la **similitud del coseno** entre el vector de la consulta y la matriz de vectores de preguntas.  
   c. Devuelva la **respuesta del √≠ndice m√°s similar**.


```python
# Instalaci√≥n e importaci√≥n de spaCy y el modelo
!pip install spacy --quiet
!python -m spacy download es_core_news_sm --quiet

import spacy
import numpy as np
# La similitud del coseno la reusamos de sklearn

print("Cargando modelo de spaCy...")
# 1. Cargar el modelo en espa√±ol
# (Deshabilitamos componentes que no usaremos para m√°s velocidad)
nlp = spacy.load('es_core_news_sm', disable=['parser', 'ner', 'tagger'])

print("Modelo cargado.")

# 2. Crear la matriz de vectores para el dataset
# Obtenemos el .vector de cada documento (pregunta)
vectores_preguntas = [nlp(pregunta).vector for pregunta in preguntas]

# Convertimos la lista de vectores a una matriz NumPy para eficiencia
matriz_vectores_preguntas = np.array(vectores_preguntas)

print("Forma de la matriz de vectores (Embeddings):", matriz_vectores_preguntas.shape)

# 3. Funci√≥n para responder usando embeddings
def responder_spacy(consulta_usuario):
    """
    Encuentra la mejor respuesta usando Embeddings de spaCy y similitud del coseno.
    """
    # 1. Obtener el vector de la consulta del usuario
    consulta_doc = nlp(consulta_usuario)
    consulta_vec = consulta_doc.vector

    # 2. Calcular la similitud del coseno
    # Necesitamos reformatear el vector de consulta (1, 96) para que coincida con la matriz (50, 96)
    similitudes = cosine_similarity(consulta_vec.reshape(1, -1), matriz_vectores_preguntas).flatten()

    # 3. Encontrar el √≠ndice de la mayor similitud
    indice_mejor_coincidencia = np.argmax(similitudes)

    # 4. Devolver la respuesta y la pregunta original
    # (Aqu√≠ no ponemos umbral, ya que los embeddings siempre tendr√°n alguna similitud)
    return respuestas[indice_mejor_coincidencia], preguntas[indice_mejor_coincidencia]

# Prueba r√°pida (opcional)
# consulta_prueba_spacy = "¬øcu√°ndo recibo mi compra?"
# respuesta_spacy, pregunta_match_spacy = responder_spacy(consulta_prueba_spacy)
# print(f"Consulta: {consulta_prueba_spacy}")
# print(f"Respuesta: {respuesta_spacy}")
# print(f"Pregunta Original (spaCy): {pregunta_match_spacy}")
```

##### Observaci√≥n del Script

Las instalaciones se ejecutaron correctamente y el modelo `es_core_news_sm` se carg√≥.  
La forma de la **matriz de vectores** es `(50, 96)`, lo que significa que tenemos **50 preguntas**,  
y cada una est√° representada por un **vector de 96 dimensiones** (el tama√±o del *embedding* de este modelo de **spaCy**).

Mi expectativa es que este chatbot `responder_spacy` sea **mejor que** `responder_tfidf` para manejar **sin√≥nimos y par√°frasis**

#### **Comparativa de Chatbots**

##### üß© Chatbot TF-IDF + Similitud del Coseno
- **Fortalezas:**  
  - Extremadamente **r√°pido de implementar y ejecutar**.  
  - Muy eficaz cuando las consultas de los usuarios utilizan las **mismas palabras clave** (o muy similares) que las preguntas del dataset.  
  - Como se observ√≥ (8/10 aciertos), si el dataset es bueno, su rendimiento es alto.  
- **Debilidades:**  
  - Es **"ingenuo" sem√°nticamente**.  
  - Falla con **sin√≥nimos no presentes** en el corpus (ej. *"dinero"* vs *"plata"*).  
  - Si el usuario formula una pregunta de manera muy distinta, el modelo se perder√°.

##### ü§ñ Chatbot Embeddings (spaCy) + Similitud del Coseno
- **Fortalezas:**  
  - Su capacidad de **entender el significado** es su mayor ventaja.  
  - Puede asociar *"cu√°ndo recibo mi compra"* con *"¬øcu√°nto tarda el env√≠o?"* aunque compartan pocas palabras.  
  - Es mejor para **clasificar preguntas nuevas** dentro del tema correcto, incluso si no tiene la respuesta exacta.  
- **Debilidades:**  
  - Es **computacionalmente m√°s costoso** (requiere cargar un modelo de ~12MB).  
  - Es susceptible a **confusiones sem√°nticas**, como el caso *remera/talles*.  
  - El modelo `es_core_news_sm` es peque√±o; uno m√°s grande (`es_core_news_lg`) probablemente funcionar√≠a mejor, aunque con mayor consumo de recursos.
